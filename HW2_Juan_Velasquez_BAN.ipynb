1) Gaussian Random Number: Gaussian random numbers. One way to generate a random number taken from the
Gaussian distribution is to use the Box-Muller formula:
𝑍 = 𝑠𝑖𝑛(2 𝜋 𝑣)(−2 𝑙𝑛 𝑢)
1
2
where u and v are real numbers between 0 and 1 generated by the random.random() function.
Compose a program that writes a standard Gaussian random variable. And calculate 𝑍 value.


```python
import math as m, random as r

u, v = r.random(), r.random()
z = m.sin(2 * m.pi * v) * m.sqrt(-2 * m.log(u))

print(f"Z value: {z}")
```

```python
import math
import random

# generate two uniform random #s between 0 & 1
U=random.random()
V=random.random()

#apply the box-muller formula
z=math.sin(2*math.pi*V)*math.sqrt(-2*math.log(U))

print(f"The calculated Z value is: {z}")
```

4) The following table contains earthquake magnitude ranges on the Richter scale and
their descriptors:
Magnitude

```python
magnitude = float(input("Enter the earthquake magnitude: "))

if magnitude < 2.0:
    descriptor = "Micro"
elif magnitude < 3.0:
    descriptor = "Very minor"
elif magnitude < 4.0:
    descriptor = "Minor"
elif magnitude < 5.0:
    descriptor = "Light"
elif magnitude < 6.0:
    descriptor = "Moderate"
elif magnitude < 7.0:
    descriptor = "Strong"
elif magnitude < 8.0:
    descriptor = "Major"
elif magnitude < 10.0:
    descriptor = "Great"
else:
    descriptor = "Meteoric"

print(f"A magnitude {magnitude} earthquake is considered to be a {descriptor} earthquake.")
```
